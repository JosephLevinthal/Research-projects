# Análise dos testes

```python
# Mostra os dados onde o z-score é maior que <threshold>

z = np.abs(stats.zscore(arquivo[['n_testes_1','n_testes_2','n_testes_3','n_testes_4','n_testes_5','n_testes_6',
                                 'n_testes_7']]))

threshold = 3
print(np.where(z > threshold))

arquivo_Outlier_Testes = arquivo[(z < threshold).all(axis=1)]
arquivo_Outlier_Testes.shape
```

```python
colunas = ['n_testes_1','n_testes_2','n_testes_3','n_testes_4','n_testes_5','n_testes_6','n_testes_7']
boxplot = arquivo_Outlier_Testes.boxplot(grid=False, rot=15, fontsize=10, column=colunas)
```

```python
colunas = ['n_submissoes_1', 'n_submissoes_2', 'n_submissoes_3', 'n_submissoes_4','n_submissoes_5',
     'n_submissoes_6', 'n_submissoes_7']
boxplot = arquivo_Outlier_Submissoes.boxplot(grid=False, rot=15, fontsize=10, column=colunas)
```
