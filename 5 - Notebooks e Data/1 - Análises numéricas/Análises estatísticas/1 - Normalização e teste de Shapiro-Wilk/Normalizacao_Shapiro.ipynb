```python
# Usaremos as bibliotecas: Pandas, Matplotlib, Scipy, Mlxtend
import pandas as pd
from matplotlib import pyplot
from scipy.stats import shapiro
from mlxtend.preprocessing import minmax_scaling, standardize
```

```python
df_completa = pd.read_excel("dadosSemDesistentes(Merge_completa).xlsx")
df_traducao = pd.read_excel("dadosSemDesistentes(Merge_traducao).xlsx")
```

# Normalização e padronização de dados
https://medium.com/data-hackers/normalizar-ou-padronizar-as-vari%C3%A1veis-3b619876ccc9

https://www.codecademy.com/articles/normalization

Normalizar (normalization) os dados normalmente é feito usando **Min-Max**

* Min-Max: garante que todos os recursos tenham exatamente a mesma escala, mas não lidam bem com outliers.

---
Padronizar (standardization) os dados normalmente é feito usando **Z-Score**

* Z-Score: lida com outliers, mas não produz dados normalizados com exatamente a mesma escala.

---

```python
# Inserir as colunas que não devem ser normalizadas
column_names_to_not_normalize = ['id_aluno', 'tipo_dica', 'sexo', 'nascimento', 'idade', 'respondeu_questionario']
column_names_to_normalize = [x for x in list(df_completa) if x not in column_names_to_not_normalize]

# Criando um dataframe com as colunas que eu quero normalizar
new_df = df_completa[column_names_to_normalize]

# Normalizando o dataframe com z-score
df_completa_padronizada = standardize(new_df, columns=column_names_to_normalize)
print(column_names_to_normalize)
df_completa_padronizada.head()

# Padronizando o dataframe com min_max
df_completa_normalizada = minmax_scaling(new_df, columns=column_names_to_normalize)
print(column_names_to_normalize)
df_completa_normalizada.head()
```

# Shapiro Wilk

## Parameters
* x: array_like (Array of sample data.)

## Returns
* W: float (The test statistic.)

* p-value: float (p-value for the hypothesis test.)
---

**O teste Shapiro-Wilk calcula se uma amostra aleatória de dados vem de uma distribuição normal. Quando o valor de p <= a 0,05 (assumindo um nível de confiança de 95%), os dados não são normais. Se esse teste falhar, você pode afirmar com 95% de confiança que seus dados não se encaixam na distribuição normal.**

---

```python
# Teste de normalidade Shapiro-Wilk, com alpha = 5%
data = df_completa["n_testes"]
stat, p = shapiro(data)
print('Statistics=%.6f, p=%.6f' % (stat, p))

# Histograma dos dados, para checar distribuição
pyplot.hist(data)
pyplot.show

# Testando se a distribuição é Gaussiana com o valor alpha definido
alpha = 0.05
confianca = 100-(alpha*100)
if p > alpha:
    print('Amostra parece Gaussiana, com %d' % (confianca) + '% de confiança (não rejeitou H0)')
else:
    print('Amostra não parece Gaussiana, com %d' % (confianca) + '% de confiança (rejeitou H0)')
```
