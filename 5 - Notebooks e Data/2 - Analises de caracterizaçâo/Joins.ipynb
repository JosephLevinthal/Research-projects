```python
import pandas as pd
```

```python
alunos_limpo = pd.read_csv("dadosSemDesistentesLIMPO.csv", sep=';')
alunos_mat = pd.read_csv("user_questionario_com_matricula_2.csv", sep=';')
alunos_cadastro_ipc = pd.read_csv("cadastro-alunos-IPC-2019-09-10.csv", sep=';', encoding='ISO-8859-1')
alunos_historico = pd.read_csv("historico-2012a2019-IPC.csv", sep=';', encoding='ISO-8859-1')
```
